```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('hauke_macroscopic (3).dat', delimiter=' ')
```

```python
Radii = data[:,0]
Masses = data[:,1]
Lambdas = data[:,2]
```

```python
plt.plot(Masses, Radii)
plt.xlabel(f"Mass ($M_\odot$)")
plt.ylabel(f"Radius (km)")
```

```python
plt.plot(Masses, Lambdas)
plt.xlabel(f"Mass ($M_\odot$)")
plt.ylabel(f"$\Lambda$")
```

```python
def mass_to_lambda(mass):
    return np.interp(mass, Masses, Lambdas)
```

```python
print(mass_to_lambda(1.3320851368494158))
print(mass_to_lambda(0.9702899110347546))
print(mass_to_lambda(1.4))
print(mass_to_lambda(0.5))
```
